```python
# gradient descent concept using named variables

'''
alpha variable is used to avoid Divergence, for when our inputs are too high, causing our prediction
values to have bigger and bigger magnitde after each iteration.
This problem is solved by having our weight mutiplied by alpha on each learning step
'''

import numpy as np

weight = 0.5
goal_prediction = 0.8
input = 2.0


alpha = 0.1

for i in range(20):
    
    pred = input * weight
    error = (pred - goal_prediction) ** 2
    
    derivative = ( input * (pred - goal_prediction) )
    weight = weight - (alpha * derivative)
    
    print("Error: " + str(error) + " Prediction:" + str(pred))
```

```python
# gradient descent with multiple inputs
# 3 inputs on this example

weights = [0.1, 0.2, 0.3]

def w_sum(a,b):
    
    assert(len(a) == len(b))
    
    output = 0
    
    for i in range(0, len(a)):
        output += a[i] * b[i]
        
    return output

def elem_multip(number, vector):
    
    output = [0, 0, 0]
    
    assert(len(output) == len(vector))
    
    for i in range(0, len(output)):
        output[i] = number * vector[i]
    
    return output
        
def neural_network(input, weights):
    
    pred = w_sum(input, weights)
    
    return pred

toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.9, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

# this is the goal_pred from other examples
win_or_lose_binary = [1, 1, 0, 1]


# representing only one input. (simulating data for the first match of a season)
true = win_or_lose_binary[0]
input = [ toes[0], wlrec[0], nfans[0] ]

# calculating prediction
pred = neural_network(input, weights)

error = (pred - true) ** 2

delta = (pred - true)

weights_delta = elem_multip(delta, weights)

alpha = 0.01

for i in range(len(weights)):
    weights[i] -= alpha * weights_delta[i]
         
    
    
    
    
    
    
    
```

```python
# debugging iterations of learning
# checking each iteration values for delta_weights and its corresponding weight adjustment

def neural_network(input, weights):
    
    out = 0
    
    for i in range(0, len(weights)):
        out += input[i] * weights[i]
        
    return out

def scalar_mult(scalar, vector):
    
    out = [0, 0, 0]
    
    for i in range(0, len(vector)):
        out[i] = scalar * vector[i]
        
    return out

toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

win_or_lose_binary = [1, 1, 0, 1]
true = win_or_lose_binary[0]

alpha = 0.01

weights = [0.1, 0.2, -.1]

input = [toes[0], wlrec[0], nfans[0]]

for i in range(5):

    pred = neural_network(input, weights)

    error = (pred - true) ** 2

    delta = pred - true

    delta_weight = scalar_mult(delta, input)
    
    print("Iteration: " + str(i+1))
    print("Pred: " + str(pred))
    print("Error: " + str(error))
    print("Delta: " + str(delta))
    print("Weights: " + str(weights))
    print("Delta Weights: " + str(delta_weight))
    
    for i in range(len(weights)):
        weights[i] = weights[i] - (alpha * delta_weight[i])
```
